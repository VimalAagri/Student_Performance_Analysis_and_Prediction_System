```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load your student performance dataset
df = pd.read_csv('student_performance_test_data.csv')
df.head()
```

```python
# Preprocessing: Encoding categorical features using LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to categorical columns
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Parental Education Level'] = label_encoder.fit_transform(df['Parental Education Level'])
df['College Type'] = label_encoder.fit_transform(df['College Type'])

# Feature selection: Only select relevant features that affect Total Percentage
X = df[['Study Hours per Day', 'Stress Level', 'Extra-Curricular Activities', 'Hindi', 'English', 'Maths', 'Science', 'Computer']]

# Target variable
y = df['Total Percentage']
```

```python


# Train a Random Forest model to see feature importance
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X, y)

# Get feature importance
feature_importance = model_rf.feature_importances_

# Display feature importance
feature_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
})

print("Feature Importance:")
print(feature_df.sort_values(by='Importance', ascending=False))

# Using the important features to train a better model
X_important = X[['Attendance Percentage', 'Study Hours per Day', 'Hindi', 'Maths', 'Science', 'Computer']]
model_rf.fit(X_important, y)

# Predict the Total Percentage
y_pred_rf = model_rf.predict(X_important)

# Save the predictions to a new CSV
df['Predicted Total Percentage'] = y_pred_rf
df.to_csv('predicted_student_performance_important_features.csv', index=False)

# Print the first few predictions
print(df[['Student_ID', 'Predicted Total Percentage']].head())
```
